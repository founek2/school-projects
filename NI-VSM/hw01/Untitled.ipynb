## Načtení dat + výpočet výskytu znaků

```python
import typing
import collections

f1 = open("../001.txt", "r")
f7 = open("../007.txt", "r")

def charOccurences(f: typing.TextIO):
    occurrences = {}
    lines = f.readlines()
    for line in lines[1:]:    # skip first line
        for k, v in collections.Counter(line).items():
            occurrences[k] = occurrences.get(k, 0) + v
    return occurrences

occurrences1 = charOccurences(f1)
occurrences7 = charOccurences(f7)
```

## Odhad pravděpodobnosti jednotlivých znaků

```python
def probabilities(occurrences: dict):
    probabilities = {}
    
    count = sum(occurrences.values())
    for k, v in occurrences.items():
        probabilities[k] = v / count
    
    return dict(sorted(probabilities.items()))

probabilities1 = probabilities(occurrences1)
probabilities7 = probabilities(occurrences7)
```

```python
from matplotlib import pyplot as plt
from scipy import stats 
import probscale

def plot_probabilities(prob: list, text: str):
    fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(15, 6))
    fig.suptitle(text)
    # Generates a probability plot of sample data against the quantiles of a specified theoretical distribution 
    # (the normal distribution by default)
    stats.probplot(prob, plot=ax1)

    ax2.plot(prob);
    ax2.set_xlabel("Znak")
    ax2.set_ylabel("Pravděpodobnost")
    
plot_probabilities(list(probabilities1.values()), "Text 1 pravděpodobnosti")
plot_probabilities(list(probabilities7.values()), "Text 7 pravděpodobnosti")
```

## Výpočet entropie

```python
from scipy.stats import entropy

entr1 = entropy(list(probabilities1.values()), base=2)
entr7 = entropy(list(probabilities7.values()), base=2)
print(f"Entropie textu 1 = {entr1}")
print(f"Entropie textu 7 = {entr7}")
```

## Optimální kód pomocí Huffmanova kódování + střední délka kódu

```python
import huffman

def calc_medium_len(code: dict, probabilities: dict):
    result = 0
    for k, v in code.items():
        prob = probabilities[k]
        result += len(v) * prob

    return result

optimal_code1 = huffman.codebook(probabilities1.items())
optimal_code7 = huffman.codebook(probabilities7.items())

med_len1 = calc_medium_len(optimal_code1, probabilities1)
med_len7 = calc_medium_len(optimal_code7, probabilities7)

print("Střední délka opt. kódu:\n" +
      f"text1 => {med_len1:.4f},\n" +
      f"text7 => {med_len7:.4f},\n\n" +
      "vymezení střední délky opt. kódu H(X) <= L(C) < H(x)+1:\n" +
      f"text1 => {entr1:.4f} <= {med_len1:.4f} < {entr1+1:.4f}\n" +
      f"text7 => {entr7:.4f} <= {med_len7:.4f} < {entr7+1:.4f}")
print("\nKodová slova pro text1:")
optimal_code1
```

## Střední délka optimálního kódu pro text1 aplikovaný na text2
### Původně když jsem načítal i první řadek
Text1 neobsahuje znaky: C,H,J. Text2 ovšem ano. 
Kódování ovšem obsahuje více kódových slov než text2 různých znaků.
Provedl jsem tedy mapování - seřadil jsem sestupně znaky dle jejich pravděpodobnosti a kódová slova dle jejich délky.Znaky s nějvětší pravděpodobností mají tak nejkratší kódová slova.

```python
def calc_medium_len_by_order(code: dict, probabilities: dict):
    sorted_code = dict(sorted(code.items(), key=lambda item: item[1], reverse=True))
    sorted_prob = dict(sorted(probabilities.items(), key=lambda item: item[1], reverse=True))

    mapper = dict(zip(sorted_prob.keys(), sorted_code.keys()))
 
    result = 0
    for k, prob in sorted_prob.items():
        v = code[mapper[k]]
        result += len(v) * prob

    return result

med_len7_using_code1 = calc_medium_len_by_order(optimal_code1, probabilities7)
print("Střední délka opt. kódování pro text1 vypočtena pro text 2")
print(f"text7 => {entr7:.4f} <= {med_len7_using_code1:.4f} ≮ {entr7+1:.4f}")
print("Optimální kód pro text1 není optimální i pro text2.\n"+
      f"Lze nalézt Huffmanovým kódováním opt. kód s kratší střední délkou = {med_len7:.4f}")
```
